```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.style as style

# Use a professional plot style
style.use('seaborn-v0_8-whitegrid')
```

```python
# Load the dataset from the data folder
file_path = 'D:\Codes\pred_maintain\data\data.csv'
df = pd.read_csv(file_path)

# Clean column names to remove trailing spaces and special characters
df.columns = df.columns.str.strip()

print("Data loaded and columns cleaned successfully.")
df.head()
```

```python
print("--- Data Info ---")
df.info()

print("\n--- Statistical Description ---")
df.describe()
```

```python
# Isolate the two decay state columns
compressor_decay_states = df['GT Compressor decay state coefficient'].unique()
turbine_decay_states = df['GT Turbine decay state coefficient'].unique()

print("--- Unique Decay States ---")
print(f"Compressor Decay States (first 10): {compressor_decay_states[:10]}")
print(f"Turbine Decay States (first 10): {turbine_decay_states[:10]}")
```

```python
# Create a figure to hold the subplots
plt.figure(figsize=(18, 12))

# Plot 1: Key Sensor Readings Over Time
plt.subplot(2, 1, 1) # (2 rows, 1 column, 1st subplot)
plt.plot(df.index, df['GT Compressor inlet air temperature (T1) [C]'], label='Compressor Inlet Temp (T1)')
plt.plot(df.index, df['GT Compressor outlet air temperature (T2) [C]'], label='Compressor Outlet Temp (T2)')
plt.plot(df.index, df['Hight Pressure (HP) Turbine exit temperature (T48) [C]'], label='HP Turbine Exit Temp (T48)')
plt.title('Key Sensor Readings Over Time', fontsize=16)
plt.xlabel('Time (index)', fontsize=12)
plt.ylabel('Temperature [C]', fontsize=12)
plt.legend()
plt.tight_layout(pad=3.0)

# Plot 2: Engine Health Decay Over Time
plt.subplot(2, 1, 2) # (2 rows, 1 column, 2nd subplot)
plt.plot(df.index, df['GT Compressor decay state coefficient'], label='Compressor Decay State', color='orange')
plt.plot(df.index, df['GT Turbine decay state coefficient'], label='Turbine Decay State', linestyle='--', color='purple')
plt.title('Engine Health Decay Over Time', fontsize=16)
plt.xlabel('Time (index)', fontsize=12)
plt.ylabel('Decay State Coefficient', fontsize=12)
plt.ylim(0.94, 1.01)  # Zoom in on the decay for better visibility
plt.legend()

# Display the plot
plt.show()
```
